```python
# Imports  
import random

from datetime import datetime
from citipy import citipy

# Import the dependencies.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Import the requests library.
import requests
# Import the API key.
from config import weather_api_key
```

```python
#Create a new set of 2,000 random latitudes and longitudes.
lats = np.random.uniform(-90.0,90.0,size=1500)
lngs = np.random.uniform(-180.0,180.0,size=1500)
lats_lngs = zip(lats, lngs)
coordinates = list(lats_lngs)
```

```python
#Get the nearest city using the citipy module
cities = []
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    # ensure no any duplicate cities
    if city not in cities:
        cities.append(city)
print(cities[:10], 'Generate', len(cities))

```

```python
url = "http://api.openweathermap.org/data/2.5/weather?units=imperial&APPID=" + weather_api_key

#Perform an API call with the OpenWeatherMap
city_data = []


print('Beginning Data Retrieval     ')
print("-----------------------------")

record_count = 1
set_count = 1

# use enumerate() method to loop index and item simutanously 
for i, city in enumerate(cities):
    if (i % 50 == 0 and i >= 50) :
        set_count +=1 # increment set count
        record_count = 1  # initialize at every beginning of set

    # build URL for API call
    city_url = url + '&q=' + city
    # Log the URL, record, and set numbers and the city.
    print(f'Processing Record {record_count} of Set {set_count} | {city}')
    # increment record count
    record_count += 1
    
    try: 
        js = requests.get(city_url).json()
        city_lat = js['coord']['lat']
        city_lon = js['coord']['lon']
        city_max_temp = js['main']['temp_max']
        city_humidity = js['main']['humidity']
        city_cloud = js['clouds']['all']
        city_wind_speed = js['wind']['speed']
        city_country = js['sys']['country']
        city_weather_description = js['weather'][0]['description']
        
        city_date = datetime.utcfromtimestamp(js['dt']).strftime('%Y-%m-%d %H:%M:%S')
        
        city_data.append({'City':city.title(),
                          'Lat':city_lat,
                          'Lng': city_lon,
                          'Max Temp':city_max_temp,
                          'Humidity':city_humidity,
                          'Cloudiness': city_cloud,
                          'Wind Speed':city_wind_speed,
                          'Country':city_country, 
                          'Current Description': city_weather_description})               
    except:
            print("City not found. Skipping...")
            pass

# Indicate that Data Loading is complete
print('-------------------------------')
print('Data Retrieval Complete        ')
print('-------------------------------')
```

```python
# Convert the array of dictionaries to a Pandas DataFrame.
city_data_df = pd.DataFrame(city_data)
city_data_df.head(10)
```

```python
city_data_df.to_csv('WeatherPy_Database.csv', index_label = 'City_ID')
```
